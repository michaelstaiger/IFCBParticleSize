```python
def summarize_ifcb_directory(directory: str) -> pd.DataFrame:
    directory = Path(directory)
    file_data = []
    global_max_roi = 0

    # ---- First pass: read files & find global max RoiType ----
    for file in sorted(directory.glob("*.csv")):
        df = pd.read_csv(file)

        if "RoiType" not in df.columns:
            continue

        roi_counts = df["RoiType"].value_counts().to_dict()
        max_roi_here = max(roi_counts.keys())
        global_max_roi = max(global_max_roi, max_roi_here)

        file_data.append((file.name, df, roi_counts))

    rows = []

    # ---- Second pass: build consistent rows ----
    for filename, df, roi_counts in file_data:

        # fill 0..global_max_roi
        roi_summary = {
            f"roi{i}": roi_counts.get(i, 0)
            for i in range(global_max_roi + 1)
        }

        vfinal = df["VolumeAnalyzed"].max()
        inhibfinal = df["InhibitTime"].max()
        runtime_final = df["RunTime"].max()
        looktime = runtime_final - inhibfinal

        row = {
            "filename": filename,
            "vfinal": vfinal,
            "inhibittime_final": inhibfinal,
            "looktime": looktime,
            **roi_summary
        }

        rows.append(row)

    return pd.DataFrame(rows)
```

### Some chat that looks like it should work to loop this over a list of PIDs that will download, ingest, summarize and save these summary stats to an output df first is just ingest IFCB from /utils/ingestIFCBData then new functions that build on it

```python


# import pandas as pd
# from pathlib import Path
# from typing import Optional, Tuple

def ingest_ifcb(
    adc_path: str,
    hdr_path: str,
    class_csv_path: Optional[str] = None,
    drop_zero_roi: bool = True,
    drop_false_trigger: bool = True,
    false_trigger_runtime_s: float = 0.25
) -> Tuple[pd.DataFrame, pd.DataFrame, Optional[pd.DataFrame]]:
    """
    Ingest IFCB ADC+HDR, optionally merge with a class CSV (if provided and exists).

    Updates:
      - Adds RoiType:
            0 if RoiX=RoiY=RoiHeight=RoiWidth=0
            else n where n is the count of rows with the same trigger#
      - Drops early false trigger(s) BEFORE drop_zero_roi
      - Recomputes InhibitTimeDiff after filtering:
            first row: InhibitTimeDiff = InhibitTime
            later rows: InhibitTimeDiff = diff(InhibitTime)
      - If class file exists: merges ADC->class (left join on RoiNumber) so zero ROIs are preserved
        and fills missing class values with 0.

    Returns:
        out_df: merged (adc + class if available) OR adc-only table
        adc_df: full ADC dataframe with derived columns
        class_df: loaded class dataframe if available else None
    """
    adc_path = Path(adc_path)
    hdr_path = Path(hdr_path)
    class_path = Path(class_csv_path) if class_csv_path else None

    # 1) Parse ADCFileFormat from .hdr
    headers = None
    with open(hdr_path, 'r') as f:
        for line in f:
            if line.startswith("ADCFileFormat:"):
                headers = [h.strip() for h in line.split(":", 1)[1].split(",")]
                break
    if not headers:
        raise ValueError(f"ADCFileFormat not found in header file: {hdr_path}")

    # 2) Load .adc with headers
    adc_df = pd.read_csv(adc_path, header=None)
    adc_df.columns = headers[:adc_df.shape[1]]

    # 3) Add RoiNumber to ADC as 1..N (do NOT change after this)
    adc_df["RoiNumber"] = range(1, len(adc_df) + 1)

    # --- NEW: RoiType (computed before any dropping) ---
    roi_cols = ['RoiX', 'RoiY', 'RoiHeight', 'RoiWidth']
    if all(c in adc_df.columns for c in roi_cols):
        for c in roi_cols:
            adc_df[c] = pd.to_numeric(adc_df[c], errors='coerce')

        is_zero = (
            (adc_df["RoiX"].fillna(0) == 0) &
            (adc_df["RoiY"].fillna(0) == 0) &
            (adc_df["RoiHeight"].fillna(0) == 0) &
            (adc_df["RoiWidth"].fillna(0) == 0)
        )

        # Count occurrences of each trigger# across ALL rows (including zeros)
        if "trigger#" in adc_df.columns:
            trig_counts = adc_df["trigger#"].map(adc_df["trigger#"].value_counts())
            trig_counts = pd.to_numeric(trig_counts, errors="coerce").fillna(1).astype(int)
        else:
            # If trigger# missing, treat each non-zero ROI as unique
            trig_counts = pd.Series(1, index=adc_df.index)

        # RoiType = 0 for zero ROIs, else count(trigger#)
        adc_df["RoiType"] = 0
        adc_df.loc[~is_zero, "RoiType"] = trig_counts.loc[~is_zero]
        adc_df["RoiType"] = adc_df["RoiType"].astype(int)
    else:
        # If ROI geometry columns missing, default: non-zero / unique
        adc_df["RoiType"] = 1

    # ---- drop early false trigger(s) BEFORE drop_zero_roi ----
    if drop_false_trigger:
        needed = set(roi_cols + ['RunTime'])
        if needed.issubset(adc_df.columns) and len(adc_df) > 0:
            adc_df['RunTime'] = pd.to_numeric(adc_df['RunTime'], errors='coerce')

            false_mask = (
                (adc_df['RunTime'] < false_trigger_runtime_s) &
                (adc_df["RoiType"] == 0)
            )

            if false_mask.any():
                adc_df = adc_df.loc[~false_mask]  # preserve RoiNumber; don't reset index

    # 4) Remove zero-ROI rows (preserving original RoiNumber values)
    if drop_zero_roi and "RoiType" in adc_df.columns:
        adc_df = adc_df.loc[adc_df["RoiType"] != 0]  # keep original RoiNumber; don't reset index

    # 5) InhibitTimeDiff (with special first-row rule after filtering)
    if 'InhibitTime' in adc_df.columns:
        adc_df['InhibitTime'] = pd.to_numeric(adc_df['InhibitTime'], errors='coerce')

        adc_df['InhibitTimeDiff'] = adc_df['InhibitTime'].diff()

        # first remaining row: InhibitTimeDiff = InhibitTime
        if len(adc_df) > 0:
            first_idx = adc_df.index[0]
            adc_df.loc[first_idx, 'InhibitTimeDiff'] = adc_df.loc[first_idx, 'InhibitTime']

        # safety: fill NaNs and clamp negatives
        adc_df['InhibitTimeDiff'] = pd.to_numeric(adc_df['InhibitTimeDiff'], errors='coerce').fillna(0.0)
        adc_df.loc[adc_df['InhibitTimeDiff'] < 0, 'InhibitTimeDiff'] = 0.0
    else:
        adc_df['InhibitTimeDiff'] = pd.NA

    # 6) VolumeAnalyzed
    if {'RunTime', 'InhibitTime'}.issubset(adc_df.columns):
        adc_df['RunTime'] = pd.to_numeric(adc_df['RunTime'], errors='coerce')
        adc_df['VolumeAnalyzed'] = (adc_df['RunTime'] - adc_df['InhibitTime']) / 240
    else:
        adc_df['VolumeAnalyzed'] = pd.NA

    # Columns to expose from ADC side
    cols_to_keep = ["trigger#", 'RoiNumber', "PMTB", 'ADCtime', 'RunTime',
                    'InhibitTime', 'InhibitTimeDiff', 'VolumeAnalyzed', 'RoiType']
    for extra in ['RoiHeight', 'RoiWidth', 'RoiX', 'RoiY']:
        if extra in adc_df.columns:
            cols_to_keep.append(extra)

    adc_out = adc_df[cols_to_keep].copy()

    # 7-8) Optional: Load class CSV + merge (ADC-driven to preserve zero ROIs)
    class_df = None
    if class_path and class_path.exists():
        class_df = pd.read_csv(class_path)

        if 'pid' not in class_df.columns:
            raise ValueError(f"Expected 'pid' column in class CSV to extract RoiNumber: {class_path}")

        class_df['RoiNumber'] = class_df['pid'].str.split('_').str[-1].astype(int)

        # ADC -> class (left) keeps all ADC rows, including zero ROIs (if not dropped)
        merged_df = adc_out.merge(class_df, on='RoiNumber', how='left')

        # Fill missing class values with 0 (e.g., zero ROIs)
        class_cols = [c for c in class_df.columns if c not in ['pid', 'RoiNumber']]
        if class_cols:
            merged_df[class_cols] = merged_df[class_cols].fillna(0)

        return merged_df, adc_df, class_df

    # No class file: return ADC-derived table as the "merged" output
    return adc_out, adc_df, None



```

```python
# from __future__ import annotations

# from dataclasses import dataclass
# from pathlib import Path
# from tempfile import TemporaryDirectory
# from typing import Iterable, Optional, Dict, Any, List

# import pandas as pd
# import requests




def _download_file(url: str, out_path: Path, timeout: int = 120) -> None:
    out_path.parent.mkdir(parents=True, exist_ok=True)
    with requests.get(url, stream=True, timeout=timeout) as r:
        r.raise_for_status()
        with open(out_path, "wb") as f:
            for chunk in r.iter_content(chunk_size=1024 * 1024):
                if chunk:
                    f.write(chunk)


def _summarize_ingested_df(df: pd.DataFrame, max_roi_type: int = 7) -> Dict[str, Any]:
    # ROI counts (0..max_roi_type)
    counts = df["RoiType"].value_counts().to_dict() if "RoiType" in df.columns else {}
    roi_summary = {f"roi{i}": int(counts.get(i, 0)) for i in range(max_roi_type + 1)}

    vfinal = pd.to_numeric(df["VolumeAnalyzed"], errors="coerce").max()
    inhibfinal = pd.to_numeric(df["InhibitTime"], errors="coerce").max()
    runtime_final = pd.to_numeric(df["RunTime"], errors="coerce").max()
    looktime = runtime_final - inhibfinal

    return {
        **roi_summary,
        "vfinal": float(vfinal) if pd.notna(vfinal) else None,
        "inhibittime_final": float(inhibfinal) if pd.notna(inhibfinal) else None,
        "looktime": float(looktime) if pd.notna(looktime) else None,
    }


def summarize_bins_streaming(
    dashboard_url: str,
    pids: Iterable[str],
    *,
    drop_zero_roi: bool = False,          # IMPORTANT: keep zeros if you want roi0 counts
    drop_false_trigger: bool = True,
    false_trigger_runtime_s: float = 0.25,
    max_roi_type: int = 7,
    session: Optional[requests.Session] = None,
) -> pd.DataFrame:
    """
    For each pid, download adc+hdr, ingest, summarize, discard files, repeat.
    Returns a DataFrame with one row per pid.

    dashboard_url examples:
      - "https://habon-ifcb.whoi.edu/<dataset>/"  (often works)
      - "https://ifcb-data.whoi.edu/<dataset>/"   (often works)

    Download pattern follows: f"{dashboard_url.rstrip('/')}/{pid}.adc" and ".hdr".
    """
    sess = session or requests.Session()
    base = dashboard_url.rstrip("/")

    rows: List[Dict[str, Any]] = []

    for pid in pids:
        adc_url = f"{base}/{pid}.adc"
        hdr_url = f"{base}/{pid}.hdr"

        try:
            with TemporaryDirectory() as tmpdir:
                tmpdir = Path(tmpdir)
                adc_path = tmpdir / f"{pid}.adc"
                hdr_path = tmpdir / f"{pid}.hdr"

                _download_file(adc_url, adc_path)
                _download_file(hdr_url, hdr_path)

                # ---- your existing ingester ----
                out_df, adc_df, class_df = ingest_ifcb(
                    adc_path=str(adc_path),
                    hdr_path=str(hdr_path),
                    class_csv_path=None,
                    drop_zero_roi=drop_zero_roi,
                    drop_false_trigger=drop_false_trigger,
                    false_trigger_runtime_s=false_trigger_runtime_s,
                )

                summary = _summarize_ingested_df(out_df, max_roi_type=max_roi_type)
                summary["pid"] = pid
                rows.append(summary)

        except Exception as e:
            # keep going; store error for later inspection
            rows.append({"pid": pid, "error": str(e)})

    return pd.DataFrame(rows)
```
