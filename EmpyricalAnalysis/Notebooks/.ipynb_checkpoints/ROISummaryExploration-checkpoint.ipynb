```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../IFCBData/summaries/nausetSum.csv")
```

```python
print(df.head())
print(df.columns.tolist())
```

```python
## Split pid into new date time and ifcb columns
# Split on "_"
split_cols = df["pid"].str.split("_", expand=True)

df["datetime_str"] = split_cols[0]
df["ifcb"] = split_cols[1]
df["datetime"] = pd.to_datetime(
    df["datetime_str"].str[1:],  # remove "D"
    format="%Y%m%dT%H%M%S"
)

## Sort by datetime
df = df.sort_values("datetime")
```

```python
print(df.head())
print(df.columns.tolist())
```

```python
## Calculate inhibit/look raton
df["inhibitRatio"]= df["inhibittime_final"]/df["looktime"]
```

```python
## Looks like one crazy value my guess is it is a bad sample so i am going to filter just the 99percentile of data
upper = df["inhibitRatio"].quantile(0.99)
df_filtered = df[df["inhibitRatio"] <= upper]
```

```python
plt.figure()
plt.plot(df_filtered["datetime"], df_filtered["inhibitRatio"])
plt.xlabel("Datetime")
plt.ylabel("InhibitTime / LookTime")
plt.title("Inhibit Fraction Over Time")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure()
plt.boxplot(df_filtered["inhibitRatio"])
plt.xlabel("Nauset")
plt.ylabel("Inhibit Ratio")
#plt.xlim(1000)
plt.tight_layout()
plt.show()
```
